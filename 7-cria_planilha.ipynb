```python
import pandas as pd
```

```python
dados = {
    'TIMES': ['Flamengo', 'Palmeiras', 'São Paulo', 'Santos', 'Corinthians'],
    'PARTIDAS JOGADAS': [10,10,10,10,10],
    'VITÓRIAS': [7,6,5,4,3],
    'EMPATES': [2,3,1,3,2],
    'Derrotas': [1,2,3,3,5],
    'GOLS_PRO': [22,18,16,14,10],
    'GOLS_CONTRA': [10,12,15,16,20],
    
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df['SALDO_GOLS'] = df['GOLS_PRO'] - df['GOLS_CONTRA']
df
```

```python
df = df.drop('PONTOS', axis=1)
```

```python
df['PONTOS'] = df['VITÓRIAS'] *3 + df['EMPATES'] * 1
```

```python
df = df.sort_values(by='PONTOS', ascending=False)
df
```

```python
df.to_excel("dados/campeonato_futebas.xlsx", index=False)
```
